### Standard Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
import sys
sys.path.append("../exp2/")
%load_ext autoreload
%autoreload 2
%matplotlib inline
```

### Read Data

```python
df = pd.read_csv("../data/beijingb_scaled.csv", index_col = 0)
df = df.rename(columns={'ts': 'Time', 'station_id': 'Station'})
```

### Import the Class 

```python
from qbc_random import ActiveLearning
```

```python
stations = df['Station'].unique()
```

```python
learners = {i:KNeighborsRegressor(n_neighbors=i+3) for i in range(5)}
```

```python
from qbc_random import ActiveLearning
al = ActiveLearning(
    df = df,
    train_stations = list(stations[:6]),
    pool_stations = list(stations[6:30]),
    test_stations = list(stations[30:]),
    learners = learners,
    context_days = 30,
    frequency = 30,
    test_days = 150,
    train_days = 100,
    number_of_seeds = 10,
    number_to_query = 1
            )
al.querybycommittee()
al.random_sampling()
```

```python
qbc_error = al.qbc_rmse[~np.isnan(al.qbc_rmse)]
random_error = al.random_rmse.mean(axis=0)[~np.isnan(al.random_rmse.mean(axis=0))]
win_loss = (qbc_error < random_error).sum() / qbc_error.shape[0]

print(qbc_error.mean(), random_error.mean(), win_loss * 100)
```
