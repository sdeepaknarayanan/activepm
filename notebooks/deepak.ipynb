```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(columns=["i", 'j', "reg", "X", "Y"])
df.head()
lastKDays = 30
```

```python
for i in range(6):
    for j in range(5):
        try:
            gp = np.load(f"../exp2/src/results/{lastKDays}/final_gp/{i}_{j}/final_rmse.npy")
            qbc = np.load(f"../exp2/src/results/{lastKDays}/final_qbc_1/{i}_{j}/final_rmse.npy")
            qbc_rd = np.load(f"../exp2/src/results/{lastKDays}/final_random_qbc_1/{i}_{j}/final_rmse.npy")
            gp_rd = np.load(f"../exp2/src/results/{lastKDays}/final_random_gp/{i}_{j}/final_rmse.npy")
            for reg, temp in zip(["gp", "qbc", "qbc_rd", "gp_rd"], [gp, qbc, qbc_rd.mean(axis=0), gp_rd.mean(axis = 0)]):
                store = {
                    'i': [i]*len(list(temp)),
                    'j': [j]*len(list(temp)),
                    'reg': [reg] * len(list(temp)),
                    "X": [i for i in range(len(temp))],
                    "Y": temp,
                }
                tempdf = pd.DataFrame(store)
                df = df.append(tempdf, ignore_index=True)
            
        except Exception as e:
            print (len(gp))
            print (e)
#             raise ValueError
```
